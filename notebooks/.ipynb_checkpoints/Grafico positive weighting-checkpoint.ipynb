```python
import os
import pickle
import numpy as np
import pandas as pd
import random
import matplotlib.pyplot as plt

from PIL import Image
import math
from sklearn.metrics import accuracy_score,hamming_loss, accuracy_score, f1_score, precision_score, recall_score

%matplotlib inline
from matplotlib import pyplot as plt
import seaborn as sns

from sklearn.metrics import multilabel_confusion_matrix, ConfusionMatrixDisplay
from sklearn.metrics import classification_report

from textwrap import wrap


from utils import KunischMetrics
from utils import KunischPruner
from utils import DataExplorer
from utils import KunischPlotter
```

```python
output_dir = os.path.join('..', 'outputs', 'alexnetpw')
target = 'mr4'
target_name = '4-Match Ratio'
scores = {}

nlabels = [26]
experimentos = ['base', 'rain_ref_rot']
apellidos = ['', '_weighted_1', '_weighted_1_2', '_weighted_1_4', '_weighted_1_6',
             '_weighted_1_8', '_weighted_2_0', '_weighted_2_2']


# Armar diccionario de experimentos
for exp in experimentos:
    for apellido in apellidos:
        scores[exp + apellido] = []

# Iterar sobre archivos de resultados sacando el target buscado
for n in nlabels:
    for exp in experimentos:
        for apellido in apellidos:
            path_predicciones = os.path.join(output_dir, exp, f'{n}L{apellido}', 'metadata.csv')
            print(path_predicciones)
            predicciones = pd.read_csv(path_predicciones, index_col = 0)
            score = float(predicciones.loc[[target]].values[0][0])
            #print(path_predicciones, score[0][0])
            scores[exp + apellido].append(score)
            
```

```python
techniques = []
for exp in scores.keys():
    label = exp.replace('weighted_', 'K=')
    if 'base' in label: # base_weighted_1_2_3_4
        label = label.replace('_', ' ', 1)
    else: #rain_ref_rot_weighted_1_2_3_4
        label = label.replace('_', ' ', 3)
    label = label.replace('_', '.')
    techniques.append(label)
techniques

s_pw = []
for sc in scores.values():
    s_pw.append(sc[0])
s_pw
```

```python
plt.rcParams.update({'font.size': 11})
plt.rcParams['figure.dpi'] = 1000
```

```python
s_base = s_pw[0:8]
s_rrr = s_pw[8:]
techniques = ['0 (No P.W.)', '1', '1.2', '1.4', '1.6', '1.8', '2.0', '2.2']
assert len(s_base) == len(s_rrr) == len(techniques)
```

```python
for i in range(0, len(techniques)):
    print(techniques[i], s_base[i], s_rrr[i])
    
# wrap de las etiquetas cuando exceden cierto largo
techniques = ['\n'.join(wrap(l, 20)) for l in techniques]

fig, ax = plt.subplots()
width = 0.4
x = np.arange(len(techniques))

b1 = ax.bar(x - width/2, s_base, width, color='tab:blue', label='Dataset base')
b2 = ax.bar(x + width/2, s_rrr, width, color='tab:orange', label='Dataset rain, ref, rot')

ax.set_ylabel(target_name)
ax.set_xlabel('Valor de K')

ax.set_title(target_name + '\nAlexNet sobre 26 etiquetas más frecuentes\nVariando K en P.W.')

# linea horizontal marcando el resultado base no pw
#plt.hlines(y=0.6496, color='red', linewidth=2, xmin=-0.6, xmax=7.6, label = 'Base, sin P.W.')
#plt.hlines(y=0.4471, color='black', linewidth=2, xmin=-0.6, xmax=7.6, label = 'Base, con P.W. K=1')

# rotar labels del eje x
#plt.xticks(rotation=90)
fig.set_size_inches(8, 11)
fig.set_dpi(200)

ax.grid(which='both')
ax.minorticks_on()
# Customize the major grid
ax.grid(which='major', linestyle='-', linewidth='0.5')
# Customize the minor grid
ax.grid(which='minor', linestyle=':', linewidth='0.5')

            
# legends en orden deseado
handles, ls = plt.gca().get_legend_handles_labels()
order = [2, 3, 0, 1]
ax.legend(loc="upper left")

ax.bar_label(b1, fmt= '%.4f', padding=-40, color='white', rotation=90)
ax.bar_label(b2, fmt= '%.4f', padding=-40, color='white', rotation=90)
    
ax.set_xticks(x, techniques)
ax.set_ylim(0, 1)
plt.show()

```
