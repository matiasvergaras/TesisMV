```python
import pandas as pd
import os
import plotly.express as px
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_json(os.path.join('..', 'labels', 'original_df.json'), orient='index')
```

```python
labels_freq = df.sum(axis=0).sort_values(ascending=False)
labels_freq
```

```python
plt.rcParams.update({'font.size': 11})
plt.rcParams['figure.dpi'] = 1000
```

```python
fig, ax = plt.subplots()
width = 0.6

(N, bins, patches) = ax.hist(labels_freq, label = 'Frecuencia', color='tab:blue', alpha=0.9, bins=200)

x = []
y = []
for i in range(1, len(N)):
    x.append(bins[i])
    y.append(N[0]/i**1.4)

ax.set_yscale('log')
ax.plot(x, y, color='orange', label='Distribución según Ley de Zipf con a=1.4')
ax.set_xlabel("Frecuencia en los datos")
ax.set_ylabel("Cantidad de etiquetas")

ax.set_title('\nDistribución cantidad de etiquetas según frecuencia.')


# rotar labels del eje x
#plt.xticks(rotation=90)
fig.set_size_inches(8, 11)
fig.set_dpi(200)

ax.grid(which='both')
ax.minorticks_on()
# Customize the major grid
ax.grid(which='major', linestyle='-', linewidth='1', alpha=0.5)
# Customize the minor grid
ax.grid(which='minor', linestyle=':', linewidth='1', alpha=0.5)
ax.legend()
ax.set_ylim(0.9 , 500)

plt.show()
```
