```python
import pandas as pd
import os
```

### Tradicional RN18

```python
dir_outputs = os.path.join('..', 'outputs', 'resnet18')
dir_exp = os.path.join('26-26L')
file = 'resultados.csv'

best_methods = {}
beta = 2**2

for tecnica in os.listdir(dir_outputs):
    if 'best' in tecnica:
        continue
    path_file = os.path.join(dir_outputs, tecnica, dir_exp, file)
    df = pd.read_csv(path_file, index_col = 0)
    df['f1-beta'] = (1 + beta) * (df['precision']*df['recall'])/((beta*df['precision']) + df['recall'])
    df = df[['f1-beta']]
    im = df['f1-beta'].idxmax()
    best_row = df.loc[[im]]
    #print(tecnica)
    #display(df.loc[[im]])
    best_methods[tecnica] = {'metodo':best_row.index[0],
                            'f1-beta': best_row.values[0][0]}

best_methods_df = pd.DataFrame.from_dict(best_methods, orient='index')
display(best_methods_df)
best_methods_df.to_csv("best_methods_rn18.csv")
```

### C2AE Patterns as Features

```python
dir_outputs = os.path.join('..', 'outputs', 'C2AE_images')
dir_exp = os.path.join('26L')
file = 'metadata.csv'

scores = None
for tecnica in os.listdir(dir_outputs):
    path_file = os.path.join(dir_outputs, tecnica, dir_exp, file)
    df = pd.read_csv(path_file, index_col = 0)
    scores = pd.concat([scores, df], axis=1)
    #df = df[['hamming_score']]
    #im = df['hamming_score'].idxmax()
    #display(tecnica)
    #display(df.loc[[im]])

scores.columns = scores.iloc[0]
scores = scores.drop(scores.index[0:6])
scores = scores.transpose()
scores = scores.sort_values('mr1', ascending=False)
display(scores)
```

```python
scores.to_csv("resultados_C2AE_images.csv")
```
